```python
df2=df1.sort_values('Wickets',ascending=False).head(20)
plt.figure(figsize=(16,10 ))
plt.bar(df2['Player Name'], df2['Wickets'], color='blue')
plt.xlabel('Wickets')
plt.ylabel('Player Name')
plt.title('Number of Wickets by Bowler in ODI')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df2=df1.sort_values('4 Wicket Haul',ascending=False).head(20)
plt.figure(figsize=(16,8))
plt.bar(df2['Player Name'], df2['4 Wicket Haul'], color='skyblue')
plt.xlabel(' Wickets')
plt.ylabel('Player Name')
plt.title('Number of 4 Wicket Hauls by Bowler in ODI')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df2=df1.sort_values('5 Wicket Haul',ascending=False).head(20)
plt.figure(figsize=(16,10 ))
plt.bar(df2['Player Name'], df2['5 Wicket Haul'], color='skyblue')
plt.xlabel('Wickets')
plt.ylabel('Player Name')
plt.title('Number of 5 Wicket Hauls by Bowler in ODI')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
